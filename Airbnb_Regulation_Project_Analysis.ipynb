  <h1 style="color: black;"><strong>Finding Average Price of Airbnb's as per Localities</strong></h1>
</div>

```python
paris_listings_neighbourhood = paris_city_listings.groupby("neighbourhood").agg({"price" : "mean"}).sort_values("price",ascending=False)
```

```python
paris_listings_neighbourhood
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (15,7) )
plt.rcParams['font.family'] = 'Times New Roman'
ax = sns.barplot(data = paris_listings_neighbourhood, 
                    x = "price" , y = paris_listings_neighbourhood.index , color = "#FF5A5F")

for i, v in enumerate(paris_listings_neighbourhood["price"]):
    ax.text(v + 5, i, f"€{v:.1f}", color='black', va='center')

plt.title("Neighbourhood in Paris (Premium Localities at the Top!)")
plt.xlabel("Expensive prices! in (€)")
plt.ylabel("Airbnb's Prominent Places Localities")
plt.show()
```

<div style="font-family: 'Times New Roman'; text-align: left;">
Inside of paris, the price of Airbnb is directly proportional to popularity of the tourist place or the view of Eiffel Tower.<br>
We need to further see if prices are impacts when the regulations are imposed.
</div>

```python
paris_listings_accommodates = paris_city_listings.groupby("accommodates").agg({"price" : "mean"}).sort_values("price",ascending=False)
```

```python
paris_listings_accommodates
```

<div style="font-family: 'Times New Roman'; text-align: left;">
The Airbnb with 15 people accommodates is the expensive, even more expensive than 16 accommodates.<br>
It might be due to Luxury,or the view of Prominent Places like Eiffle Tower nearby.<br>
This might be outliers in the 15 people accommodates, which is increasing its average price.

```python
plt.figure(figsize = (15,6.5) )
plt.rcParams['font.family'] = 'Times New Roman'
ax = sns.barplot(data = paris_listings_accommodates, 
                    x = "price" , y = paris_listings_accommodates.index ,
                    order = paris_listings_accommodates.index,
                    orient = "h" ,color = "#00A699")
for i, v in enumerate(paris_listings_accommodates["price"]):
    ax.text(v + 5, i, f"€{v:.1f}", color='black', va='center')

plt.title("Accommodates in Paris (Premium Localities at the Top!)")
plt.xlabel("Expensive prices! in (€)")
plt.ylabel("Airbnb's Accommodates in Prominent Places Localities")
plt.show()
```

```python
paris_city_listings["year_of_listing"] = paris_city_listings["host_since"].dt.year
```

```python
paris_city_listings.head()
```

```python
paris_listings_over_time = paris_city_listings.groupby("year_of_listing").agg({"listing_id" :"count" , "price" : "mean"})
```

```python
paris_listings_over_time
```

```python
plt.figure(figsize = (15,6) )
plt.rcParams['font.family'] = 'Times New Roman'
sns.lineplot(data = paris_listings_over_time["listing_id"], color = "#FC642D" , linewidth=2)
plt.title("Airbnb's (Paris City) Listings Over the Time")
plt.xlabel("Year of Listings")
plt.ylabel("Airbnb's Count of Listings")
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (15,6) )
plt.rcParams['font.family'] = 'Times New Roman'
sns.lineplot(data = paris_listings_over_time["price"], color = "#767676" ,linewidth = 2)
plt.title("Airbnb's (Paris City) Listings Over the Time")
plt.xlabel("Year of Listings")
plt.ylabel("Airbnb's Prices in (€)")
plt.grid()
plt.show()
```

<div style="background: linear-gradient(to right, #f88379, #ffb199); padding: 20px; border-radius: 10px; 
            font-size: 20px; font-family: 'Times New Roman'; text-align: center;">
  <h1 style="color: black;"><strong>Finding Effects of Regulations on Airbnb's Market Over the Time</strong></h1>
</div>

```python
fig, ax = plt.subplots(figsize=(15, 6))

line1, = ax.plot(paris_listings_over_time.index, 
                 paris_listings_over_time["listing_id"], 
                 color="#FC642D", linewidth=2, label="Number of Listings")

ax2 = ax.twinx()
line2, = ax2.plot(paris_listings_over_time.index, 
                  paris_listings_over_time["price"], 
                  color="#767676", linewidth=2, label="Average Price")

ax2.set_ylim(0)

lines = [line1, line2]
labels = [line.get_label() for line in lines]
ax.legend(lines, labels, loc="upper left")

ax.set_xlabel("Date")
ax.set_ylabel("Number of Listings")
ax2.set_ylabel("Average Price (€)")
plt.title("Airbnb's (Paris City) Listings and Price Over Time")
ax.grid()
plt.tight_layout()
plt.show()
```

  <h1 style="color: black;"><strong>Final Insights</strong></h1>
</div>

<div style="font-family: 'Times New Roman'; text-align: left;">
1. In the early phase after launch, the number of Airbnbs steadily increased along with prices, driven by initial traction and interest from early adopters.<br>
2. As the platform gained widespread popularity and became a common utility, listings continued to grow, but prices began to decline gradually.<br>
3. Following the announcement of regulations around 2015, confidence in the business dipped—resulting in a decline in the no of listings,while prices started increasing.<br>
4. By 2019, once the regulations had become the new norm, the number of Airbnbs began rising again, and prices started to fall due to the increased supply.
</div>

  <h1 style="color: black;"><strong>Recommendation for Airbnb's Market</strong></h1>
</div>

<div style="font-family: 'Times New Roman'; text-align: left;">
1. Regulations on long-term rentals can negatively affect the business, potentially leading to customer and host churn due to increased uncertainty.<br>
2. Such regulatory measures may contribute to Airbnb’s financial losses, which could be challenging to recover in the future.<br>
3. If these changes begin to impact customer experience, it could result in a distorted or negative perception of the brand.<br>
4. It is advisable to closely monitor similar regulatory developments in other regions and proactively prepare for them.<br>
5. Airbnb can consider implementing similar regulatory practices in other markets as a proactive measure.<br>
6. The platform can introduce stricter guidelines for onboarding and offboarding hosts to ensure compliance and quality.<br>
7. Limiting the number of Airbnb listings within a locality could help maintain balance in the rental market and prevent government intervention.
</div>
